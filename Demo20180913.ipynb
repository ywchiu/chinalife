## 檔案操作

```python
fid = open('test.txt', 'w')
fid.write('hello\nworld')
fid.close()
```

```python
with open ('test.txt', 'w') as fid:
    fid.write('hello\nworld')
```

```python
with open('test.txt', 'r') as fid:
    for line in fid:
        print(line)
```

```python
with open('test.txt', 'r') as fid:
    for line in fid:
        print(line.strip())
```

```python
a  =  '                     hello             '
print(a)
```

```python
print(a.strip())
```

```python
with open('test.txt', 'r') as fid:
    s = fid.read()
print(s)
```

```python
k = 0
with open('test.txt', 'r') as fid:
    for line in fid:
        k = k + 1
print(k)
```

```python
ary = []
with open('test.txt', 'r') as fid:
    for line in fid:
        ary.append(line)
len(ary)
```
